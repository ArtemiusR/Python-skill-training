```python
import pandas as pd
```

```python
def print_target_results():
    f = pd.read_csv('targets_log.csv')
    return f
```

```python
f = print_target_results().sort_values(by='delta', ascending=False)
f.info()
```

```python
f.sort_values(by='delta', ascending=False)
```

```python
g = f.loc[f.move == 'stop_buy100','delta']
print(g.sum())
print(g.median())
print(g.mean())
print(len(g.index.tolist()))
```

```python
moves = ['buy','sell']
levels = [100, 90, 80]
for move in moves:
    for level in levels:
        g = f.loc[f.move == f'stop_{move}{level}','delta']
        print(f'{move} {level}')
        print('Sum: ', g.sum())
        print('Median: ', g.median())
        print('Mean: ', g.mean())
        print('Entries: ', len(g.index.tolist()))
        if g.mean() >= 0:
            print('Potential: ', (g.mean() - 0.03) * len(g.index.tolist()), '\n')
        else:
            print('Potential: ', (g.mean() + 0.03) * len(g.index.tolist()) * (-1), '\n')
```
